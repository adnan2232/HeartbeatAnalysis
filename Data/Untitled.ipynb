```python
import numpy as np
import pandas as pd
import librosa
from librosa import display
from IPython.display import Audio
from os.path import isfile
```

```python
def fetch_csv_data_without_na(path):
    frame = pd.read_csv(path, names=["dataset","WAV","category","sublabel"])
    frame.drop(["sublabel","dataset"], axis=1, inplace=True)
    frame.dropna(inplace=True)
    frame = frame[1:]
    frame.reset_index(drop=True,inplace=True)
    return frame
```

```python
def name_formatting_setB(setB):
    return setB["WAV"].map(lambda x:str(x).replace("Btraining_","")\
                              .replace("extrastole_","extrastole__")\
                              .replace("normal_","normal__").replace("murmur_","murmur__")\
                              .replace("murmur__noisymurmur__","murmur_noisymurmur_")\
                              .replace("normal__noisynormal__","normal_noisynormal_"))
```

```python
setA = fetch_csv_data_without_na("set_a.csv")
setB = fetch_csv_data_without_na("set_b.csv")
```

### Execute below block only once

```python
setB["WAV"] = name_formatting_setB(setB)

for x in setB["WAV"]:
    if not isfile(x):
        print(f"file {x} not found")
        break
```
